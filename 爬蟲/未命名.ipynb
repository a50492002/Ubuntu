```python
twses = twstock.twse
```

```python
twses_data = pd.DataFrame()

for key in twses:
    data = twses[key]
    if data.CFI == 'ESVUFR':
        twselist = [data.name, data.code, data.ISIN, data.start, data.market, data.group, data.CFI]
        df = pd.DataFrame([twselist],columns=['name', 'code','ISIN','start','market','group','CFI'])
        twses_data=pd.concat([twses_data,df],axis=0,ignore_index=True)
twses_data
```

```python
sql="INSERT INTO Stocks_No(`Number`, `Name`, `Country`, `Market`, `Group`, `Start`) VALUES "
for i in range(len(twses_data)):
    updata=twses_data.loc[i]
    if twses_data.loc[0]['code'] != updata['code']:
        sql+=','
    sql+="('%s', '%s', '%s', '%s', '%s', '%s')" % (updata['code'], updata['name'], 'TW', updata['market'], updata['group'], updata['start'])
sql
```

```python
try:
    # 輸入 SQL指令
    cursor.execute(sql)
    # 執行 SQL指令
    conn.commit()
except:
    # 錯誤時重新執行
    conn.rollback()
```
